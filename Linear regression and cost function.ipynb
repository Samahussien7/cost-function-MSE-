```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
x_val=[2,4,6,8,10,12]
y_val=[5,6,6,8,5,12]
```

```python
x=np.array(x_val,float)
y=np.array(y_val,float)
x1=np.linspace(0,13)
y1=x1
y2=.5*x1
y3=.5*x1+4
plt.plot(x_val,y_val,"o")
plt.plot(x1,y1,"red")
plt.plot(x1,y2,"blue")
plt.plot(x1,y3,"yellow")
plt.show()
```

```python
def costfun(x,y,w,b):
    total=0.0
    y_p=w*x+b
    total=0.5*(np.mean((y_p-y)**2))
    return total
```

```python
c1=costfun(x,y,1,0)
c1
```

```python
c2=costfun(x,y,.5,0)
c2
```

```python
c3=costfun(x,y,.5,4)
c3
```

```python

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
w=[1,.5,.5]
b=[0,0,4]
cost=[c1,c2,c3]
ax.plot(w,b,cost,color = "red")
plt.show()
```
